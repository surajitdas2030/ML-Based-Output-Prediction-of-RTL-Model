```python
df = pd.read_csv('shift_32bit.txt',sep=" ",header=None)
col_str_dic = {column:str for column in list(df)}
df = pd.read_csv("shift_32bit.txt",sep=" ",header=None, dtype=col_str_dic)
x_clk = df.iloc[:,0]
x_bit = df.iloc[:,1]
y_gen = df.iloc[:,2]
x_samples=[]
y_samples=[]
k = 0
for x in x_bit:
  k = k+1

temp = list(x_clk[0])
temp_list=[]
for c in temp:
  temp_list.append(int(c))
temp = list(x_bit[0])
for c in temp:
  temp_list.append(int(c))
temp = list(y_gen[0])
for c in temp:
  temp_list.append(int(c))
x_samples.append(temp_list)

for i in range(1,k):
  temp = list(x_clk[i])
  temp_list=[]
  for c in temp:
    temp_list.append(int(c))
  temp = list(x_bit[i])
  for c in temp:
    temp_list.append(int(c))
  temp = list(y_gen[i-1])
  for c in temp:
    temp_list.append(int(c))
  x_samples.append(temp_list)


for y in y_gen:
  # print(y)
  temp = list(y)
  temp_list=[]
  for c in temp:
    temp_list.append(int(c))
  y_samples.append(temp_list)

print(x_samples[2])
print(y_samples[2])
```

```python
x_temp = np.array(x_s) # 16 inputs
x_t = []
for a in x_temp:
  temp = a.flatten()
  x_t.append(temp)
x = np.array(x_t)
print(x[2])
y_temp = np.array(y_s) # 16 inputs
y_t = []
for a in y_temp:
  temp = a.flatten()
  y_t.append(temp)
y = np.array(y_t)
print(y[2])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True)
#x_train, x_test, y_train, y_test = train_test_split(x_samples, y_samples, test_size=0.3, shuffle=True)
```

```python
#ANN1
#model = models.Sequential(name='Parity_checker')
#model.add(layers.Input(shape=(32,), name='input'))
#model.add(layers.Dense(units=32,activation=activations.relu, name='hidden1'))
## model.add(layers.Dense(units=3, activation=activations.sigmoid, name='hidden2'))
#model.add(layers.Dense(units=1, activation=activations.sigmoid, name='output'))

#add the layers
#ANN2:0.99 accuracy
# model = tf.keras.models.Sequential()
# model.add(tf.keras.layers.Dense(64, input_dim=32, activation='relu'))
# model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# ANN3:0.99 accuracy
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(34, input_dim=34, activation='relu'))
model.add(tf.keras.layers.Dense(34, activation='relu'))
model.add(tf.keras.layers.Dense(34, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
#model.add(tf.keras.layers.Dense(9, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='sigmoid'))

# model = tf.keras.models.Sequential()
# model.add(tf.keras.layers.Dense(32, input_dim=16, activation='relu'))
# # model.add(tf.keras.layers.Dense(32, activation='relu'))
# model.add(tf.keras.layers.Dense(16, activation='relu'))
# model.add(tf.keras.layers.Dense(8, activation='relu'))
# model.add(tf.keras.layers.Dense(4, activation='relu'))
# model.add(tf.keras.layers.Dense(2, activation='relu'))
# model.add(tf.keras.layers.Dense(16, activation='sigmoid'))

model.summary()
```
