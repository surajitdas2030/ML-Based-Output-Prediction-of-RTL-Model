```python
import numpy as np

import tensorflow as tf
from tensorflow.keras import models, layers, activations
import random
print(tf.__version__)
np.random.seed(0)
tf.random.set_seed(0)
```

```python
num_examples = 20000
num_classes = 32

#construct input values ,data set stored in the file: input_output_32_bit.txt 
def input_values():
    multiple_values = [map(int, '{0:016b}'.format(i)) for i in range(2 ** 16)]
    f = open("input_output_32_bit.txt", "w")
    random.shuffle(multiple_values)
    final_values = []
    for value in multiple_values[:num_examples]:
        temp = []
        f.write("[")
        count = 0
        for number in value:
            temp.append([number])
            count^=number
            f.write(str(number)+",")
        f.write("],\t")
        f.write("["+str(count)+"]\t"+"\n")
        final_values.append(temp)
    f.close
    return final_values
    
#calculate output values
def output_values(inputs):
    final_values = []
    for value in inputs:
        # output_values = [0 for _ in range(num_classes)]
        count = 0
        for i in value:
            count ^= i[0]
        # if count < num_classes:
            # output_values[count] = 1
        final_values.append([count])
    return final_values
```

```python
x_samples = input_values()
x_samples2 = input_values()
y_samples = output_values(x_samples)
y_samples2 = output_values(x_samples2)
```
