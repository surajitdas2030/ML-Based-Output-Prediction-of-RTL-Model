```python
#HandShake Testing Data *****************************
df = pd.read_csv('2.txt',sep=" ",header=None)
col_str_dic = {column:str for column in list(df)}
df = pd.read_csv("2.txt",sep=" ",header=None, dtype=col_str_dic)
x0 = df.iloc[:,0]
x1 = df.iloc[:,1]
k = 0
for x in x0:
  k = k+1

#new data start

x3_samples=[]
y3_samples=[]
y4_samples=[]

for i in range(0,k-2):
  temp = list(x0[i])
  temp_list=[]
  temp_list_y=[]
  if x0[i]=="1" and x0[i+1]=="0":
    #print(x0[i],x1[i])
    #print(x0[i+1],x1[i+1]) 
    #print("Falling edge of Valid")
    y4_samples.append(1)
  else:
    y4_samples.append(0)
    
  for c0 in temp:
    temp_list.append(int(c0))
    '''
    if c0=="1":
      print("c0=",c0)
    else:
      print("c0!=1")
    '''
    #temp_list_y.append(int(c0))
  temp = list(x1[i])
  for c1 in temp:
    temp_list.append(int(c1))
  x3_samples.append(temp_list)
  #y3_samples.append(temp_list_y)

for i in range(1,k-1):
  temp = list(x0[i])
  temp_list_y=[]
  for c0 in temp:
    temp_list_y.append(int(c0))
  y3_samples.append(temp_list_y)

print("*********New Data***")
len(x3_samples)
print("x3_samples=",len(x3_samples),x3_samples)
print("y3_samples=",len(y3_samples),y3_samples)
print("y4_samples=",len(y4_samples),y4_samples)
print("************")
#numpy array start
x_temp = np.array(x3_samples) 
x_t = []
for a in x_temp:
  temp = a.flatten()
  x_t.append(temp)
x4 = np.array(x_t)
print("x4=",len(x4),x4)

y5_temp = np.array(y4_samples) 
y5_t = []
for a in y5_temp:
  temp = a.flatten()
  y5_t.append(temp)
y5 = np.array(y5_t)
print("y5=",len(y5),y5)

y_temp = np.array(y3_samples) 
y_t = []
for a in y_temp:
  temp = a.flatten()
  y_t.append(temp)
y4 = np.array(y_t)
print("y4=",len(y4),y4)
#numpy array end

#new data end
```

```python

from sklearn.model_selection import train_test_split
#x1_train, x1_test, y1_train, y1_test = train_test_split(x4, y4, test_size=0.3, shuffle=False)
x1_train, x1_test, y1_train, y1_test = train_test_split(x4, y5, test_size=0.3, shuffle=False)

#x_train, x_test, y_train, y_test = train_test_split(x_samples, y_samples, test_size=0.3, shuffle=True)
```

```python
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Dense(4, input_dim=2, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
model1.add(tf.keras.layers.Dense(2, activation='relu'))
model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model1.summary()
```

```python
#model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
model1.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
#model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
```

```python
model1.fit(x1_train, y1_train, epochs=100)
score = model1.evaluate(x1_test, y1_test, verbose=2)
```
