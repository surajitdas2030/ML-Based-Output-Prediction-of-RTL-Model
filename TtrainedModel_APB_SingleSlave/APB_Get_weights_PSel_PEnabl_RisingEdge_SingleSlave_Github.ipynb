```python

from sklearn.model_selection import train_test_split
#x1_train, x1_test, y1_train, y1_test = train_test_split(x4, y4, test_size=0.3, shuffle=False)
x1_train, x1_test, y1_train, y1_test = train_test_split(x4, y4, test_size=0.3, shuffle=False)

#x_train, x_test, y_train, y_test = train_test_split(x_samples, y_samples, test_size=0.3, shuffle=True)
```

```python
layers=2
inputs=2
neurons_l0=2
neurons_l1=1
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Dense(2, input_dim=2, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
#model.add(tf.keras.layers.Dense(4, activation='relu'))
#model1.add(tf.keras.layers.Dense(2, activation='relu'))
model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model1.summary()
```

```python
#model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
model1.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
#model.compile(loss='mse', optimizer='adam',metrics=['accuracy'])
weight1 = model1.get_weights()
print("weight1")
print(weight1)
```
